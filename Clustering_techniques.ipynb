K-Means Clustering

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/content/sample_data/Mall_Customers.csv")
df.head()
```

```python
df.describe
```

```python
plt.scatter(df.Age,df['Annual_Income_(k$)'])
plt.xlabel('Age')
plt.ylabel('Annual_Income_(k$)')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Annual_Income_(k$)']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
plt.plot(km.cluster_centers_)
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Annual_Income_(k$)'],color='green')
plt.scatter(df2.Age,df2['Annual_Income_(k$)'],color='red')
plt.scatter(df3.Age,df3['Annual_Income_(k$)'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('Annual_Income_(k$)')
plt.legend()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Annual_Income_(k$)']])
df['Annual_Income_(k$)'] = scaler.transform(df[['Annual_Income_(k$)']])

scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
```

```python
df.head()
```

```python
plt.scatter(df.Age,df['Annual_Income_(k$)'])
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Annual_Income_(k$)']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Annual_Income_(k$)'],color='green')
plt.scatter(df2.Age,df2['Annual_Income_(k$)'],color='red')
plt.scatter(df3.Age,df3['Annual_Income_(k$)'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age','Annual_Income_(k$)']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

Hierarchical clustering
Single Link, Complete Link and Average Link

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import numpy as np

linked = linkage(df[['Age','Annual_Income_(k$)']], 'single')

labelList = range(1, 201)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

```python
#Complete linkage
linked = linkage(df[['Age','Annual_Income_(k$)']], 'complete')

labelList = range(1, 201)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()

```

```python
#Average linkage
linked = linkage(df[['Age','Annual_Income_(k$)']], 'average')

labelList = range(1, 201)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

DB Scan clustering 

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Determine centroids
centers = [[0.5, 2], [-1, -1], [1.5, -1]]
#Create dataset
X, y = make_blobs(n_samples=400, centers=centers, 
                  cluster_std=0.5, random_state=0)
#Normalize the values
X = StandardScaler().fit_transform(X)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='Paired')
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.4, min_samples=20)
db.fit(X)
```

```python
y_pred = db.fit_predict(X)
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1],c=y_pred, cmap='Paired')
plt.title("Clusters determined by DBSCAN")
```
